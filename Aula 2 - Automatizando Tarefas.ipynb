- Passo 2: Gerar as análises solicitadas pelo gestor de forma automática
- Passo 3: Enviar um e-mail automáticamente para o gestor com os resultados das análises

# Passo 1

```python
!pip install yfinance
```

```python
import yfinance
```

```python
ticker = input("Digite a sigla da ação desejada: ")
dados = yfinance.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

# Passo 2

Análises dos últimos 6 meses:

- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

import time
time.sleep(15)

pyautogui.click(x=98, y=170)

pyperclip.copy("eloisa.c@outlook.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy("Análises Diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""Prezado, 

segue análise dos últimos seis meses da ação {ticker}:

Cotação mínima: R$ {round(maxima, 2)}
Cotação máxima: R$ {round(minima, 2)}
Cotação atual: R$ {round(atual, 2)}

Qualquer dúvida, estou a disposição!
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=1265, y=1001)
```
